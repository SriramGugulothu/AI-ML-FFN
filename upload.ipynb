```python
from polyglot.text import Text
```

```python
def transliterate_to_korean(input_file, output_file): # this is for conversion of engish to korean
    # Read input file in utf-8 format
    with open(input_file, 'r', encoding='utf-8') as f:
        english_text = f.read()

    # Created text object 
    text_obj = Text(english_text)

    # Transliterate to Korean 
    korean_text = text_obj.transliterate(target_language='ko') # based on inupt it will judge it is english to transliterate

    # Korean_text is not string, its a list.So converting it back to string.
    korean_text_str = ' '.join(korean_text)

    # Writing transliterated text to output file in utf-8 format
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(korean_text_str)
```

```python
def transliterate_to_english(input_file, output_file):
    # Read input file in utf-8 format
    with open(input_file, 'r', encoding='utf-8') as f:
        korean_text = f.read()

    # Created text object 
    text_obj = Text(korean_text)

    # Transliterate to Korean 
    en_text = text_obj.transliterate(target_language='en')

    # Korean_text is not string, its a list.So converting it back to string.
    en_text_str = ' '.join(en_text)

    # Writing transliterated text to output file in utf-8 format
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(en_text_str)
```
